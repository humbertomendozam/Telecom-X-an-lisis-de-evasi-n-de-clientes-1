```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)

distritos = [
"San Isidro",
"Miraflores",
"Surco",
"La Molina",
"San Borja",
"Jesús María",
"Lince",
"Magdalena",
"Los Olivos",
"San Juan de Lurigancho"
]

n = 100

data = {
"cliente_id": range(1,n+1),
"distrito": np.random.choice(distritos,n),
"compras": np.random.randint(1,20,n),
"gasto_total": np.random.randint(200,5000,n),
"meses_cliente": np.random.randint(1,36,n),
"reclamos": np.random.randint(0,5,n)
}

df = pd.DataFrame(data)

df["abandono"] = np.where(
(df["compras"] < 5) & (df["reclamos"] > 2),
"Si",
"No"
)

df.head()
```

```python
df.to_csv("clientes_lima.csv", index=False)
```

```python
abandono = df["abandono"].value_counts()

print(abandono)
```

```python
abandono.plot(kind="bar")

plt.title("Evasión de Clientes en Lima")
plt.xlabel("Estado")
plt.ylabel("Cantidad")

plt.show()
```

```python
tabla = pd.crosstab(df["distrito"], df["abandono"])

print(tabla)
```
